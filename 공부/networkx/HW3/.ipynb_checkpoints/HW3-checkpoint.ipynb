```python
import networkx as nx
import pandas as pd
import sklearn
from sklearn.cluster import SpectralClustering
import networkx as nx
import matplotlib.pyplot as plt
```

```python
import wikipediaapi
wiki1 = wikipediaapi.Wikipedia(language='ko', extract_format=wikipediaapi.ExtractFormat.WIKI) 
p_wiki1 = wiki1.page("미세먼지") 
print("Page - Exists: %s" % p_wiki1.exists())
wiki2 = wikipediaapi.Wikipedia(language='ko', extract_format=wikipediaapi.ExtractFormat.WIKI) 
p_wiki2 = wiki2.page("스모그") 
print("Page - Exists: %s" % p_wiki2.exists())
```

```python
from konlpy.tag import Twitter
from collections import Counter
nlpy1 = Twitter()
nouns1 = nlpy1.nouns(p_wiki1.text)
print(nouns1)
```

```python
count = Counter(nouns1)
tag_count = []
tags = []
for n, c in count.most_common():
    dics = {'tag': n, 'count': c}
    if len(dics['tag']) >= 2 and len(tags) <= 49:
        tag_count.append(dics)
        tags.append(dics['tag'])
for tag in tag_count:
    print(" {:<14}".format(tag['tag']), end='\t')
    print("{}".format(tag['count']))
```

```python
nlpy2 = Twitter()
nouns2 = nlpy2.nouns(p_wiki2.text)
print(nouns2)
```

```python
count = Counter(nouns2)
tag_count = []
tags = []
for n, c in count.most_common():
    dics = {'tag': n, 'count': c}
    if len(dics['tag']) >= 2 and len(tags) <= 49:
        tag_count.append(dics)
        tags.append(dics['tag'])
for tag in tag_count:
    print(" {:<14}".format(tag['tag']), end='\t')
    print("{}".format(tag['count']))
```

```python
G=nx.Graph()
G1=nx.Graph()
```

```python
text1=[]
text1=p_wiki1.text.split(".")
text2=[]
text2=p_wiki2.text.split(".")
```

```python
i=0
for each_line in text1:
#for i in range (len(nouns1)):
    if each_line.find(nouns1[i])>0:
        nlpy3 = Twitter()
        nouns3 = nlpy3.nouns(each_line)
        G.add_nodes_from(nouns3)
        for j in range (len(nouns3)):
            for l in range (len(nouns3)):
                if nouns3[j]!=nouns3[i]:
                    print(nouns3[j],nouns3[i])
                    G.add_edges_from([(nouns3[j],nouns3[i])])
                    G.add_edges_from([(nouns3[i],nouns3[j])])
                    break
```

```python
i=0
for each_line in text2:
#for i in range (len(nouns1)):
    if each_line.find(nouns2[i])>0:
        nlpy4 = Twitter()
        nouns4 = nlpy4.nouns(each_line)
        G1.add_nodes_from(nouns4)
        for j in range (len(nouns4)):
            for l in range (len(nouns4)):
                if nouns4[j]!=nouns4[i]:
                    print(nouns4[j],nouns4[i])
                    G1.add_edges_from([(nouns4[j],nouns4[i])])
                    G1.add_edges_from([(nouns4[i],nouns4[j])])
                    break
```

```python
kern=nx.algorithms.community.kernighan_lin_bisection(G)
print(kern)
```

```python
kern1=nx.algorithms.community.kernighan_lin_bisection(G1)
print(kern1)
```

```python
adj_matrix = nx.to_numpy_matrix(G) 
node_list = list(G.nodes()) 
clusters = SpectralClustering(affinity = 'precomputed', assign_labels="discretize",random_state=0,n_clusters=2).fit_predict(adj_matrix)

for i in range (len(clusters)):
    print(node_list[i],clusters[i])
```
